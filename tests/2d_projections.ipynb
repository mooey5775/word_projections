```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xaxis = np.random.rand(300).tolist()
yaxis = np.random.rand(300).tolist()
```

```python
xaxis
```

```python
basis = np.c_[xaxis, yaxis]
```

```python
basis
```

```python
basis_inv = np.linalg.pinv(basis)
```

```python
plt.scatter(*basis_inv.dot(np.random.rand(300, 10000)))
```
